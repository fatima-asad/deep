```python
import cv2
import pytesseract
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = Image.open(list(uploaded.keys())[0])

# Resize (important for low resources)
image = image.resize((800, 800))

# Convert to OpenCV
img_cv = np.array(image)
gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)

# OCR
text = pytesseract.image_to_string(gray)

print("----- EXTRACTED TEXT -----")
print(text)
```

```python
with open("extracted_menu.txt", "w") as f:
    f.write(text)
```

```python
files.download("extracted_menu.txt")
```

```python
import re

def clean_menu_text(raw_text):
    lines = raw_text.split("\n")
    cleaned = []

    for line in lines:
        line = line.strip()
        if len(line) < 5:
            continue

        line = line.replace("1", "i").replace("0", "o")
        line = re.sub(r"\s+", " ", line)

        cleaned.append(line)

    return cleaned
```

```python
cleaned_lines = clean_menu_text(text)

for item in cleaned_lines:
    print(item)
```

```python
def categorize_dish(dish):
    dish = dish.lower()

    if "coffee" in dish or "water" in dish or "chocolate" in dish:
        return "Drinks"
    elif "fries" in dish or "vada" in dish:
        return "Starters"
    else:
        return "Main Course"
```

```python
menu_data = []

for dish in cleaned_lines:
    item = {
        "dish": dish.title(),
        "price": "Not Available",
        "category": categorize_dish(dish)
    }
    menu_data.append(item)

for item in menu_data:
    print(item)
```

```python
import pandas as pd

df = pd.DataFrame(menu_data)
df
```

```python
df.to_csv("menu_output.csv", index=False)
```
