```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Embedding,SimpleRNN,Dense
```

```python
with open(r"/content/_MConverter.eu_poem.csv", 'r') as file:
    lines = file.readlines()
df = pd.DataFrame(lines, columns=['poem_line'])
```

```python
df.head()
```

```python
poem=df['poem_line']
```

```python
poem=poem.str.lower()
```

```python
poem.head()
```

```python
import string
translator=str.maketrans(string.punctuation,' '*len(string.punctuation))
poem=poem.apply(lambda x:x.translate(translator))
```

```python
poem.head()
```

```python
vocab_size=500
tokenizer=Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(poem)
```

```python
word_index=tokenizer.word_index
print(len(word_index))
list(word_index.items())[:10]
```

```python
sequences=tokenizer.texts_to_sequences(poem)
```

```python
X=[]
Y=[]
for sequence in sequences:
  for i in range(1,len(sequence)):
    input_seq=sequence[:i]
    output_seq=sequence[i]
    X.append(sequence[:i])
    Y.append(sequence[i])
```

```python
max_len=max([len(x) for x in X])
```

```python
import tensorflow as tf
X_padded=tf.keras.preprocessing.sequence.pad_sequences(X,maxlen=max_len,padding='pre')
```

```python
y=np.array(Y)
```

```python
from tensorflow.keras.utils import to_categorical
y_categorical=to_categorical(y,num_classes=vocab_size+1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense,LSTM,GRU
```

```python
rnn_model=Sequential()
rnn_model.add(Embedding(input_dim=vocab_size+1,output_dim=100,input_length=max_len-1))
rnn_model.add(SimpleRNN(units=100,activation='relu',return_sequences=False))
rnn_model.add(Dense(units=vocab_size+1,activation='softmax'))
```

```python
rnn_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
gru_model=Sequential()
gru_model.add(Embedding(input_dim=vocab_size+1,output_dim=100,input_length=max_len-1))
gru_model.add(GRU(units=100,activation='relu',return_sequences=False))
gru_model.add(Dense(units=vocab_size+1,activation='softmax'))
```

```python
gru_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
gru_model.summary()
```

```python
gru_model.save('gru_model.h5')
```

```python
gru_model.fit(X_padded,y_categorical,epochs=10,batch_size=32,validation_split=0.1)
```

```python
index_to_word={}
for word,index in word_index.items():
  index_to_word[index]=word
```

```python
index_to_word
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def predictor(model,tokenizer_instance,text,max_len):
   text=text.lower()
   text=text.translate(translator)
   seq=tokenizer_instance.texts_to_sequences([text])
   seq=pad_sequences(seq,maxlen=max_len,padding='pre')
   pred=model.predict(seq,verbose=0)
   pred_index=np.argmax(pred)
   return index_to_word[pred_index]
```

```python
seed_next='the cut'
next_word=predictor(gru_model,tokenizer,seed_next,max_len)
print(next_word)
```

```python
def generate(model,tokenizer,seed_next,max_len,n_words):
  for _ in range(n_words):
    next_word=predictor(model,tokenizer,seed_next,max_len)
    if next_word=="end":
      break
    seed_next+=f" {next_word}"
  return seed_next
```

```python
seed_next="the cut"
gen_word=generate(gru_model,tokenizer,seed_next,max_len,n_words=10)
print(gen_word)
```
