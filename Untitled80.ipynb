```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Embedding,SimpleRNN,Dense
```

```python
df=pd.read_csv('/content/Quotes Dataset.csv')
```

```python
df.head()
```

```python
quotes=df['Quote']
```

```python
quotes.head()
```

```python
quotes=quotes.str.lower()
```

```python
quotes.head()
```

```python
import string
translator=str.maketrans(' ',' ',string.punctuation)
quotes=quotes.apply(lambda x:x.translate(translator))
```

```python
quotes.head()
```

```python
vocab_size=1000
tokenizer=Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(quotes)
```

```python
word_index=tokenizer.word_index
print(len(word_index))
list(word_index.items())[:10]
```

```python
sequences=tokenizer.texts_to_sequences(quotes)
```

```python
for i in range (10):
  print(quotes[i])
  print(sequences[i])
```

```python
X=[]
Y=[]
for sequence in sequences:
  for i in range(1,len(sequence)):
    input_sequence=sequence[:i]
    output_sequence=sequence[i]
    X.append(sequence[:i])
    Y.append(sequence[i])
```

```python
max_len=max([len(x) for x in X])
print(max_len)
```

```python
import tensorflow as tf
X_padded=tf.keras.preprocessing.sequence.pad_sequences(X,maxlen=max_len,padding='pre')
```

```python
X_padded
```

```python
y=np.array(Y)
```

```python
from tensorflow.keras.utils import to_categorical
y_one_hot=to_categorical(y,num_classes=vocab_size)
```

```python
y_one_hot.shape
```

```python
X_padded.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense,LSTM
```

```python
Embedding_dim=50
rnn_unite=128
```

```python
rnn_model=Sequential()
rnn_model.add(Embedding(vocab_size,output_dim=Embedding_dim,input_length=max_len))
rnn_model.add(SimpleRNN(rnn_unite))
rnn_model.add(Dense(vocab_size,activation='softmax'))
```

```python
rnn_model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
rnn_model.summary()
```

```python
lstm_model=Sequential()
lstm_model.add(Embedding(vocab_size,output_dim=Embedding_dim,input_length=max_len))
lstm_model.add(LSTM(rnn_unite))
lstm_model.add(Dense(vocab_size,activation='softmax'))
```

```python
lstm_model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
lstm_model.summary()
```

```python
history_lstm=lstm_model.fit(X_padded,y_one_hot,epochs=10,batch_size=32,validation_split=0.2)
```

```python
lstm_model.save('lstm_model.h5')
```
